Mining Complete dataset 

```python
comp_df = pd.read_csv("C:/Users/shali/OneDrive/Desktop/wiki/Data_mining/Data_mining_2_updated/complete_context_withlabels.csv",index_col = 0, delimiter=";")
comp_df
```

```python
min_support = 7/len(comp_df)
# compute the frequent itemsets using fpgriowth from mlxtend
from mlxtend.frequent_patterns.fpgrowth import fpgrowth
frequent_itemsets = fpgrowth(comp_df, min_support=min_support, use_colnames = True)

# print the frequent itemsets
frequent_itemsets
```

```python
# Compute the association rules based on the frequent itemsets
from mlxtend.frequent_patterns import association_rules

# compute and print the association rules
result = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)
result
```

```python
# compute the antecedent length
result["antecedent_len"] = result["antecedents"].apply(lambda x: len(x))
result
```

```python
#selecting based on the rules
result[ (result['antecedent_len'] == 3) &
       (result['confidence'] > 0.75) &
       (result['lift'] > 1.2) ]
```

```python
df_subset = result.sample(n=50)
df_subset
```
